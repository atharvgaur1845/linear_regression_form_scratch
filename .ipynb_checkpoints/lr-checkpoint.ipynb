```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Task1.csv")
X = df.iloc[:, :-1].values  
y = df.iloc[:, -1].values
print(df.head())
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Data Points')

plt.xlabel("X (Feature)")
plt.ylabel("Y (Target)")
plt.title("Scatter Plot of X vs Y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weight = None
        self.bias = None
        self.training_history = []
    
    def fit(self, X, y):
        X = X.reshape(-1, 1)
        y = y.reshape(-1)
        n_samples = X.shape[0]
        
        #initialization
        self.weight = 0
        self.bias = 0
        
        for i in range(self.n_iterations):
           
            y_pred = (self.weight * X).reshape(-1) + self.bias
            
            # gradient
            error = y_pred - y
            dw = (1/n_samples) * np.sum(X.reshape(-1) * error)
            db = (1/n_samples) * np.sum(error)

            self.weight -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            mse = np.mean((y - y_pred) ** 2)
            self.training_history.append(mse)
    
    def predict(self, X):
        X = X.reshape(-1, 1)
        return (self.weight * X).reshape(-1) + self.bias
```

```python
def normalize_features(X):
    return (X - X.mean()) / X.std()
X_normalized = normalize_features(X)
```

```python
model = LinearRegression(learning_rate=0.01, n_iterations=1000)
model.fit(X_normalized, y)

y_pred = model.predict(X_normalized)


mse = np.mean((y - y_pred) ** 2)

print(f"Weight: {model.weight:.4f}")
print(f"Bias: {model.bias:.4f}")
print(f"Mean Squared Error: {mse:.4f}")

plt.figure(figsize=(12, 5))

#history plot
plt.subplot(1, 2, 1)
plt.plot(model.training_history)
plt.title('Training History')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.grid(True)

# line plot
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue', label='Data points')
X_sorted = np.sort(X)
X_sorted_norm = (X_sorted - X.mean()) / X.std()
y_pred_line = model.predict(X_sorted_norm)
plt.plot(X_sorted, y_pred_line, color='red', label='Regression line')
plt.title('Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
